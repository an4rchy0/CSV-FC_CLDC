```python
import pandas as pd

try :
    df = pd.read_csv('./Data/2021socio_economic_indonesia.csv')
    print(df)
except Exception as e:
    print(e)
```

Clear data

```python
missing_values = df.isnull().sum()
print(missing_values)

df = pd.DataFrame(df)
```

Menyortir Pendapatan GDP mulai dari terendah: ds1 (data sortir1)

```python
ds1 = df.sort_values(by='exp_percap')
ds1
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for pr, d in df.groupby('poorpeople_percentage'):
    ax.scatter(d['cities_reg'], d['poorpeople_percentage'], label=pr)

plt.title('Poor Region Rank')
plt.xlabel('Cities Region')
plt.ylabel('Poor Percentage (%)')
plt.grid(True)
ax.tick_params(axis='x', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```

```python
import numpy as np
import sklearn
from sklearn import preprocessing

x_train = np.array(df[['reg_gdp', 'life_exp', 'avg_schooltime']])
y_train = np.array(df['poorpeople_percentage'])

lb = preprocessing.LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_train = y_train.flatten()

print(f'x_training : \n {x_train}')
print(f'y_training : \n {y_train}')
```
